## Features Encoding

In this section we would like to encode the data by manipulating the strings data types, thus increasing the ML algorithms success rate.

### Label Encoding

In this method we convert the strings labels into a suitable set of numeric values.

```python
# pre encoding
labelEncodedData = data.copy()
for column in labelEncodedData.columns:
  print(column, 'unique values:', labelEncodedData[column].unique())
```

```python
# label encoding
labelEncoding = preprocessing.LabelEncoder()

for column in labelEncodedData.columns:
  labelEncodedData[column]= labelEncoding.fit_transform(labelEncodedData[column])
  print(column, 'unique values:', labelEncodedData[column].unique())
```

As we can see, all features now contain only numeric values, which are based on the original strings data

### One-Hot Encoding

In this method we convert our categorical labels into a set of boolean vectors.
The disadvantage on this method is that the dataframe can rapidly increase in the case of a feature with many possible values.

```python
# pre encoding
oneHotEncodingData = data.copy()
for column in oneHotEncodingData.columns:
  print(column, 'unique values:', oneHotEncodingData[column].unique())
```

```python
# one hot encoding 'class' column
oneHotEncodingData = pd.get_dummies(oneHotEncodingData)
print(oneHotEncodingData)
```

# Model Development

## Splitting the data

```python
# split the label encoding df to features and response
X_label_data = labelEncodedData.drop('class', axis = 1)
y_label_data = labelEncodedData['class']

# split the one hot encoding df to features and response
X_oh_data = oneHotEncodingData.drop(['class_e', 'class_p'], axis = 1)
y_oh_data = oneHotEncodingData[['class_e', 'class_p']]

#split the label df to train, test and validation
X_label_train, X_label_test, y_label_train, y_label_test = train_test_split(X_label_data, y_label_data, test_size = 0.3)
X_label_test, X_label_val, y_label_test, y_label_val = train_test_split(X_label_test, y_label_test, test_size = 0.5)

#split the df to train, test and validation
X_oh_train, X_oh_test, y_oh_train, y_oh_test = train_test_split(X_oh_data, y_oh_data, test_size = 0.3)
X_oh_test, X_oh_val, y_oh_test, y_oh_val = train_test_split(X_oh_test, y_oh_test, test_size = 0.5)

print(len(X_label_train), 'train examples')
print(len(X_label_test), 'test examples')
print(len(X_label_val), 'validation examples')
```

## KNN

Fit KNN with n = 5

```python
knn = KNeighborsClassifier(n_neighbors=5)

# label encoding
knn.fit(X_label_train, y_label_train)
print('Label Encoding:')
print('Validation: ', knn.score(X_label_val, y_label_val))
print('Train: ', knn.score(X_label_train, y_label_train))
print('Test: ', knn.score(X_label_test, y_label_test))

print()

# one hot
knn.fit(X_oh_train, y_oh_train)
print('One Hot:')
print('Validation: ', knn.score(X_oh_val, y_oh_val))
print('Train: ', knn.score(X_oh_train, y_oh_train))
print('Test: ', knn.score(X_oh_test, y_oh_test))
```

We can see that for KNN, the One Hot results are slightly better.

## Decision Tree

```python
tree = DecisionTreeClassifier()

# label encoding
tree.fit(X_label_train, y_label_train)
print('Label Encoding:')
print('Validation: ', tree.score(X_label_val, y_label_val))
print('Train: ', tree.score(X_label_train, y_label_train))
print('Test: ', tree.score(X_label_test, y_label_test))

print()

# one hot
tree.fit(X_oh_train, y_oh_train)
print('One Hot:')
print('Validation: ', tree.score(X_oh_val, y_oh_val))
print('Train: ', tree.score(X_oh_train, y_oh_train))
print('Test: ', tree.score(X_oh_test, y_oh_test))
```

For the decision tree model, both label encoding and one hot results are the same.
